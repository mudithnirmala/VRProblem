```python
def writeToFile(N,disMat,timeMat):

    lines = []
    with open('out.txt','w+') as file:
        lines.append(str(N)+'\n')

        for i in range(N):
            lines.append(' '.join(map(str,disMat[i]))+'\n')
            
        for i in range(N):
            lines.append(' '.join(map(str,timeMat[i]))+'\n')

        file.writelines(lines)     

```
